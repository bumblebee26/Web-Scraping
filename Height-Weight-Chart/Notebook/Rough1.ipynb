```python
title=[]
ht_wt=[]
a = len(chart_content)
```

```python
for i in range(1, a-1):
    chart = chart_content[1].find_all("a")
    b = len(chart)
    for j in range(0, b-1):
            t=chart_content[i].find_all("a")[j]
            tt = t['href']
            title = str(title) + ", " + str(tt)
            image = t.find("img")
            text = image.get('title')
            if text is None:
                continue
            else:
                ht_wt = str(ht_wt) + ", " + str(text)
```

```python
ht_wt
```

```python
title
```

```python
print(len(title.split(",")))
```

```python
print(len(ht_wt.split(",")))
```

```python
title = title.split(",")
```

```python
ht_wt = ht_wt.split(",")
```

```python
df = pd.DataFrame(columns = {"Image_link", "Height & Weight"})
df["Image_link"] = title
df["Height & Weight"] = ht_wt
```

```python
df
```

```python
df = df.loc[1:]
df
```

```python
df.to_csv("Output.csv")
```

```python
temp = link[15]
temp
```

```python
response = requests.get(temp)
soup = BeautifulSoup(response.text, 'html.parser')
img_tags = soup.find_all("img")
img_tags = img_tags[2]
img_tags
```

```python
type(img_tags)
```

```python
url = img_tags.get('src')
url
```

```python
filename = re.search(r'/([\w_-]+[.](jpg))$', url)
filename
```

```python
with open(filename.group(1), 'wb') as f:
        if 'http' not in url:
            # sometimes an image source can be relative 
            # if it is provide the base url which also happens 
            # to be the site variable atm. 
            url = '{}{}'.format(path, url)
        response = requests.get(url)
        f.write(response.content)
```
